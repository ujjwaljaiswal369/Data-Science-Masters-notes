## Handling Outliers

## 5 number Summary
1. Minimum Value
2. Q1- 25 percentile
3. Median
4. Q3- 75 percentile
5. MAximum

```python
import numpy as np
```

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74,1000,1100,1300,2020]
```

```python
# Lower fence --------- Higher fence
Q1 = np.percentile(lst_marks, [25])
print(Q1)
```

```python
minimum,Q1,Q2,Q3,maximum = np.quantile(lst_marks, [0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,Q2,Q3,maximum 
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
Lower_fence = Q1-1.5*(IQR)
Higher_fence = Q3+1.5*(IQR)
```

```python
Lower_fence,Higher_fence 
```

```python
outliers = []
for i in lst_marks :
    if i >= -12.0 and i<= 164.0 :
        print('This element is not a outlier')
        
    else :
        outliers.append(i)
```
