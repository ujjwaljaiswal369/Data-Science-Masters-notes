```python
x,y
```

```python
colurs =  np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x,y, c = colurs, s = sizes, alpha=.8, )
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('this is title')
```
